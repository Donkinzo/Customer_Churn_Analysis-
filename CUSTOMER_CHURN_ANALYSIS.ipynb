```python
#HYPOTHESIS QUERY

from scipy import stats

# define the two groups based on churn and non churned customers
group1 = df_encoded[df_encoded['Churn_Yes'] == 1]['TotalCharges']
group2 = df_encoded[df_encoded['Churn_No']==1]['TotalCharges']
# perform a t-test
t_statistic, p_value = stats.ttest_ind(group1,group2)
#set significance level alpha
alpha = 0.05
#check if the p-value is less than alpha
if p_value < alpha:
    print('Reject the null hypothesis')
else:
    print('fail to reject the null hypothesis')
```

# CONCLUSION

1. Gender Churn Distribution: The pie chart clearly displays the proportion of churned customers based on gender. it shows the churn distribution between males and females. A larger portion of the pie corresponds to "female," it means that a higher percentage of female customers churned compared to male customers. 

2. Gender-Based Churn Comparison: The Barplot chart displays two bars for each gender: one representing the number of unchurned customers and the other representing the number of churned customers.
The x-axis shows the gender categories (Female and Male), while the y-axis represents the count of customers. By comparing the heights of the "no_churn" and "churn" bars for each gender, we can visually see the churn rate for each gender.
This visualization allows us to directly compare the churn and non-churn counts for both genders, making it easy to see which gender has a higher proportion of churned customers. The male gender had a slightly more churn figure than the female gender.

3. Churn Comparison between Young Lads and Senior Citizens: The chart displays two bars for each category: one representing the number of unchurned customers ("no_churn") and the other representing the number of churned customers ("churn").
The x-axis shows the two categories: "Young lads" and "Senior Citizen."
The y-axis represents the count of customers.

This visualization showcases the churn rates between non-senior citizens ("Young lads") and senior citizens, which provides insights into whether age plays a significant role in customer churn. From the visualizaton we can agree that the Young lads are bringing in more business than the senior citizens. This barplot chart helps us understand how senior citizen status relates to customer churn and hence requires strategies aimed at retaining customers from these two age groups.

4. Churn Comparison by Contract Type: The chart displays two bars for each contract type: one representing the number of unchurned customers ("no_churn") and the other representing the number of churned customers ("churn").
The x-axis shows the three contract types: "Month to Month," "One Year," and "Two Year."
The y-axis represents the count of customers.
By comparing the heights of the "no_churn" and "churn" bars for each contract type, we can visually assess the churn rate for customers with different contract durations.
This visualization allows us to directly compare the churn rates between customers on different contract types, providing insights into whether contract duration plays a significant role in customer churn.

This information indicates a need to tailor churn mitigation strategies for customers on different contract types. For example, Month-to-Month customers have a higher churn rate, we may want to focus on strategies to incentivize longer-term contracts.
Overall, this barplot chart helps us understand how contract type relates to customer churn and can inform strategies aimed at retaining customers with varying contract durations.

5. Churned vs. Unchurned Customer Distribution: The pie chart displays two slices: one representing the number of churned customers ("Churn") and the other representing the number of unchurned customers ("Non Churn").
The labels "Churn" and "Non Churn" correspond to the two categories.
The percentages inside the slices ("autopct") indicate the proportion of churned and unchurned customers relative to the total number of customers.

Insights:

This visualization provides a high-level overview of the overall churn rate, showing what percentage of customers have churned and what percentage have not.
It helps stakeholders quickly understand the churn status of the customer base.
we can use this information to assess the overall churn rate and set benchmarks for churn mitigation efforts.

6. Relationship between Total Charges and Monthly Charges: Positive Correlation: The scatter plot shows a generally positive correlation between "Monthly Charges" and "Total Charges." As "Monthly Charges" increase, "Total Charges" tend to increase as well. This suggests that customers who pay higher monthly fees tend to accumulate higher total charges over time.

-Spread of Data: The data points are spread out across the plot, indicating some variation in the relationship. While there is a positive trend, there are also instances where customers with similar monthly charges have different total charges. This suggests that other factors may influence total charges.

-Outliers: There appear to be outliers in the data where customers with relatively low monthly charges have relatively high total charges (top-right corner of the plot). These outliers may represent high-value customers who have accumulated significant charges over time.

-Potential Churn Insights: For customer churn mitigation strategies, understanding this relationship can be valuable. Customers with higher "Monthly Charges" might be more valuable to the business in terms of "Total Charges," and efforts to retain these high-paying customers could be prioritized.

-Segmentation: Segmenting customers based on their "Monthly Charges" can help tailor retention strategies. Customers with low "Monthly Charges" but high "Total Charges" might have unique needs or reasons for high expenditures, and addressing those could help in retaining them.
